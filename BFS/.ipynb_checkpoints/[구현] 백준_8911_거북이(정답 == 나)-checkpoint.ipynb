```python
import sys

def cal_max_and_min(x, y):
  global max_x, min_x, max_y, min_y
  max_x = max(max_x, x)
  min_x = min(min_x, x)
  max_y = max(max_y, y)
  min_y = min(min_y, y)

t = int(sys.stdin.readline().strip())
direction = ['E', 'W', 'S', 'N']

for _ in range(t):
  max_x, min_x, max_y, min_y = 0, 0, 0, 0
  x, y = 0, 0
  curr_dir = 'N'
  arr = list(map(str, sys.stdin.readline().strip()))

  for c in arr:
    if curr_dir == 'N' and c == 'F':
      y += 1
      cal_max_and_min(x, y)
    elif curr_dir == 'N' and c == 'B':
      y -= 1
      cal_max_and_min(x, y)
    elif curr_dir == 'N' and c == 'L':
      curr_dir = 'W'
    elif curr_dir == 'N' and c == 'R':
      curr_dir = 'E'
    elif curr_dir == 'E' and c == 'F':
      x += 1
      cal_max_and_min(x, y)
    elif curr_dir == 'E' and c == 'B':
      x -= 1
      cal_max_and_min(x, y)
    elif curr_dir == 'E' and c == 'L':
      curr_dir = 'N'
    elif curr_dir == 'E' and c == 'R':
      curr_dir = 'S'
    elif curr_dir == 'S' and c == 'F':
      y -= 1
      cal_max_and_min(x, y)
    elif curr_dir == 'S' and c == 'B':
      y += 1
      cal_max_and_min(x, y)
    elif curr_dir == 'S' and c == 'L':
      curr_dir = 'E'
    elif curr_dir == 'S' and c == 'R':
      curr_dir = 'W'
    elif curr_dir == 'W' and c == 'F':
      x -= 1
      cal_max_and_min(x, y)
    elif curr_dir == 'W' and c == 'B':
      x += 1
      cal_max_and_min(x, y)
    elif curr_dir == 'W' and c == 'L':
      curr_dir = 'S'
    elif curr_dir == 'W' and c == 'R':
      curr_dir = 'N'
  print((max_x - min_x)*(max_y - min_y))
```

```python
import sys

N = int(sys.stdin.readline().strip())
K = int(sys.stdin.readline().strip())
arr = [[0]*N for _ in range(N)]
for _ in range(K):
  apple_x, apple_y = map(int, sys.stdin.readline().split())
  arr[apple_x-1][apple_y-1] = 1

L = int(sys.stdin.readline().strip())
direction = {}
for _ in range(L):
  tmp_direction = list(map(str, sys.stdin.readline().split()))
  direction[int(tmp_direction[0])] = tmp_direction[1]

x, y = 1, 1
time = 0
dir_index = 1
tail_length = 1
tail = [(0,0)]

while 1 <= x <= N and 1 <= y <= N and (x, y):
  

  if arr[x-1][y-1] == 1:
    tail_length += 1
    tail.append((x,y))

  if time in direction:
    if direction[time] == 'D':
      dir_index += 1
      if dir_index == 4:
        dir_index = 0 
    elif direction[time] == 'L':
      dir_index -= 1
      if dir_index == -1:
        dir_index = 3

  if dir_index == 0: #북
    x -= 1
  elif dir_index == 1: #동
    y += 1
  elif dir_index == 2: #남
    x += 1
  elif dir_index == 3: #서
    y -= 1

  time += 1

  if (x, y) in tail:
    break

  tail.append((x, y))
  del tail[0]


print(time)
```
